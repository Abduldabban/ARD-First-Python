```python
nomiric = df[['credit_amount', 'age', 'existing_credits', 'num_dependents']]
categorical = df[['credit_history', 'purpose', 'housing', 'job']]
```

4. Preprocessing

Perform any needed preprocessing on the chosen features, including:
    
• Scaling

• Encoding

Dealing with NaN values

Encode the class column

```python
from sklearn.preprocessing import LabelEncoder
df['class'] = LabelEncoder().fit_transform(df['class'])
df
```

```python
plt.title('Credit Raring Histogram')
plt.xlabel('Class')
plt.ylabel('Number of Entries')
plt.hist(df['class'], bins =2, color = 'blue')
```

Encode Categorical variables

```python
categorical = pd.get_dummies(categorical)
categorical
```

```python
credit = pd.concat([nomiric, categorical], axis = 1)
credit
```

5. Splitting the Data

Split your data as follows:

80% training set

10% validation set

10% test set

```python
y = df['class'].values

X_columns = len(credit.columns)

X = credit.iloc[:, 0:X_columns].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
print(len(X_train), len(X_test))
```

```python
from sklearn.model_selection import train_test_split
X_validate, X_test, y_validate, y_test = train_test_split(X_test, y_test, test_size=0.5, random_state=0)
```

```python
print(len(X_train), len(X_test), len(X_validate))
```

6. Training Classifiers

Use the KNN-classifier model to train your data.

```python
from sklearn.neighbors import KNeighborsClassifier
# Choose value of K
K=3
model=KNeighborsClassifier(n_neighbors=K)
model.fit(X_train, y_train)

model.score(X_test,y_test)
```

Choose the best k for the k-nearest neighbor (KNN) algorithm by trying different values and validating performance on the validation set.

Note: choosing the best k is an example of hyper-parameter tuning.

Classification Metrics

```python
scores = []

n = range(1,20)

for i in n:
    model=KNeighborsClassifier(n_neighbors=i)
    model.fit(X_train, y_train)
    result = model.score(X_test,y_test)
    scores.append(result)
    
print(scores)
```

```python
plt.plot(n,scores)
plt.show()
```

seems from both the score print and the line plot k= 13 provides the best result = 0.69

```python
K=13
model=KNeighborsClassifier(n_neighbors=K)
model.fit(X_train, y_train)

model.score(X_test,y_test)
```

Print the accuracy score of your final classifier based on the validation data since k was selected based on the test data,
so need to valisate with new set of data

```python
result = model.score(X_validate,y_validate)
print('Accuracy = ',result)
```

Print the confusion matrix.

```python
from sklearn.metrics import confusion_matrix

y_predict = model.predict(X_validate)

print('Confusion Matrix \n', confusion_matrix(y_validate, y_predict))
```
