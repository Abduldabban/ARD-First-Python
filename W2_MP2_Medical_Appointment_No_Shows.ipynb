```python
# Before extracting the requested features, let's asses how valid is the data
# Handicap is a True or False column - as per the documentation provided in Kaggle. Having values 2, 3 & 4 is not expected
#  Below desnity histogram shows 2, 3 and 4 probability and how small those compared to 0 and 1 values

plt.title('Handicap Density Histogram')
plt.xlabel('Handcap')
plt.ylabel('Number of Entries')
plt.hist(df['Handcap'], bins =4, density = True)
```

```python
# Assume all Handicap values larger than 1 as outliers - only 200 entries out of 110527 
# Eliminate the outliers
df.drop(df[df.Handcap > 1].index, inplace = True)
df
```

```python
# extracted_data is the extracted version of the original dataset, which includes only the features identified above

extracted_data = df[['Gender','Age', 'Scholarship', 'Hipertension', 'Diabetes', 'Alcoholism', 'Handcap', 'SMS_received']]
extracted_data
```

4. Preprocessing

Perform any needed preprocessing of the chosen features, including: \
scaling. \
Encoding, and \
dealing with NaN values. \
Hint: use only the preprocessing steps you believe are useful.

```python
# Encode the target column No-show

from sklearn.preprocessing import LabelEncoder
df['No-show'] = LabelEncoder().fit_transform(df['No-show'])
df.head()
```

```python
# Normalize the Age column

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
extracted_data[['Age']] = scaler.fit_transform(extracted_data[['Age']])
```

```python
#Encode the Gender column with an Ordinal Encoder

from sklearn.preprocessing import OrdinalEncoder
encoder=OrdinalEncoder()
extracted_data[['Gender']] = encoder.fit_transform(extracted_data[['Gender']])
```

```python
extracted_data.head()
```

5. Splitting the Data

Split your data as follows: \
80% training set \
10% validation set \
10% test set

```python
y = df['No-show'].values

X_columns = len(extracted_data.columns)

X = extracted_data.iloc[:, 0:X_columns].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
print(len(X_train), len(X_test))
```

```python
from sklearn.model_selection import train_test_split
X_validate, X_test, y_validate, y_test = train_test_split(X_test, y_test, test_size=0.5, random_state=0)
print(len(X_train), len(X_test), len(X_validate))
```

6. Training Tree-based Classifiers

Use a decision tree classifier model to train your data.

```python
#using a Decision Tree Classifier with default parameters

from sklearn.tree import DecisionTreeClassifier

model_tree = DecisionTreeClassifier(random_state = 0)

model_tree.fit(X_train, y_train)

tree_score = model_tree.score(X_test,y_test)
print('Decision Tree Classifier Accuracy = ', tree_score)
```

Choose the best criterion for the decision tree algorithm by trying different values and validating performance on the validation set. \
Note: choosing the best criterion is an example of hyper-parameter tuning.


```python
# Optimize Decision Tree Classifier using GridSearch

from sklearn.model_selection import GridSearchCV

# Number of trials were made on different param_grid values. Finalized on the below with max three values in each row
param_grid_tree = {'max_depth':[4, 5, 7],
              'min_samples_leaf':[10, 15, 20],
              'min_samples_split':[1, 2, 3],
              'criterion':['gini','entropy']}

grid_tree = GridSearchCV(estimator=model_tree, param_grid=param_grid_tree, 
                         verbose=True, scoring='accuracy')
grid_tree.fit(X_train, y_train)
```

```python
print('The best parameters are %s with a score of %0.2f' 
      % (grid_tree.best_params_, grid_tree.best_score_))
```

Classification Metrics \
Print the accuracy score of your final classifier. \
Print the confusion matrix.

```python
#Now we have everything in grid, including the best one, so we can just use that:

tree_score = grid_tree.score(X_validate,y_validate)
print('Decision Tree Classifier Accuracy with best criteria = ', tree_score)

y_tree_predict = grid_tree.predict(X_validate)

from sklearn.metrics import confusion_matrix
print('Decision Tree Classifier Confusion Matrix with best criteria \n', confusion_matrix(y_validate, y_tree_predict))
```

7. Random Forest

Repeat step 6. \
Increase/decrease the number of estimators in random forest and comment on the difference of the classification metrics.

```python
#using a Random Forest Classifier with default parameters

from sklearn.ensemble import RandomForestClassifier

forest_scores = []

n = range(10,250, 50)

for i in n:
    model_forest = RandomForestClassifier(n_estimators = i, criterion = 'gini', max_depth = 5, min_samples_leaf = 15, 
                                          min_samples_split = 2, bootstrap = True, random_state = 0)
    model_forest.fit(X_train, y_train)
    result = model_forest.score(X_test,y_test)
    forest_scores.append(result)

print(forest_scores)

# No noticeable difference in accuracy with different n_estimators
```

```python
#applying the same parameters on the validation data:

model_forest = RandomForestClassifier(n_estimators = 10, criterion = 'gini', max_depth = 5, min_samples_leaf = 15, 
                                      min_samples_split = 2, bootstrap = True, random_state = 0)
                                    #using the lowest tried n_estimators = 10

model_forest.fit(X_train, y_train)

forest_score = model_forest.score(X_validate,y_validate)
print('Random Forest Classifier Accuracy with untested data = ', forest_score)

y_forest_predict = model_forest.predict(X_validate)

from sklearn.metrics import confusion_matrix
print('Random Forest Classifier Confusion Matrix with untested data \n', confusion_matrix(y_validate, y_forest_predict))
```
