# 3- Splitting the Data
Split your data as follows: \
80% training set \
10% validation set \
10% test set

```python
# Use the full features dataset

y = df['diagnosis'].values

X_columns = len(df.columns)-1

X = df.iloc[:, 0:X_columns].values
```

```python
# Initial split to produce the training data

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

print(len(X_train), len(X_test))
```

```python
# Second split to produce the testing and validation data

X_validate, X_test, y_validate, y_test = train_test_split(X_test, y_test, test_size=0.5, random_state=0)
print(len(X_train), len(X_test), len(X_validate))
```

# Train Classifiers with full features dataset

KNN:

```python
from sklearn.neighbors import KNeighborsClassifier
# Choose value of K
K=5
modelKNN=KNeighborsClassifier(n_neighbors=K)
modelKNN.fit(X_train, y_train)

modelKNN.score(X_test,y_test)
```

```python
scores = []

n = range(1,20)

for i in n:
    modelKNN=KNeighborsClassifier(n_neighbors=i)
    modelKNN.fit(X_train, y_train)
    score = modelKNN.score(X_test,y_test)
    scores.append(score)
    #print(i, "gives a score of:", score)

best_score_KNN_ff = max(scores)
bestK_KNN_ff = scores.index(best_score_KNN_ff) + 1

print("KNN Classifier best score is:", best_score_KNN_ff, "with", bestK_KNN_ff, "neighbors.") 
```

```python
plt.plot(n,scores)
plt.show()
```

```python
# Validate the best score with validation data and calculate confusion matrix

modelKNN=KNeighborsClassifier(n_neighbors=bestK_KNN_ff)
modelKNN.fit(X_train, y_train)

scoreKNN_ff = modelKNN.score(X_validate,y_validate)

print('KNN classifier accuracy score with full dataset validation data and best k_neighbors = ',scoreKNN_ff)
```

```python
from sklearn.metrics import confusion_matrix

y_predict = modelKNN.predict(X_validate)

cmKNN_ff = confusion_matrix(y_validate, y_predict)

print('KNN classifier Confusion Matrix with full dataset validation data and best k_neighbors = \n', cmKNN_ff)
```

Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
modelRF = RandomForestClassifier(n_estimators = 100, criterion = 'gini', max_depth = 5, min_samples_leaf = 15, 
                                          min_samples_split = 2, bootstrap = True, random_state = 0) 
    
modelRF.fit(X_train, y_train)

modelRF.score(X_test,y_test) # with gini criterion
```

```python
# Optimize a RF classifier Model using GridSearch

param_grid={'criterion': ['gini', 'entropy'],
            'n_estimators': range(10, 150, 15),
            'max_depth': range(1, 5, 1),
            'min_samples_leaf': range(1, 5, 1),
            'min_samples_split': range(1, 5, 1)}
```

```python
# import GridSearchCV from model_selection
from sklearn.model_selection import GridSearchCV

grid = GridSearchCV(modelRF, param_grid, verbose=2)
grid.fit(X_train,y_train)
```

```python
print('The best parameters are %s with a score of %0.2f' 
      % (grid.best_params_, grid.best_score_))
```

```python
# Validate the best score with validation data and calculate confusion matrix

modelRF = RandomForestClassifier(criterion = 'entropy', max_depth = 4, min_samples_leaf = 1, 
                                 min_samples_split = 2, n_estimators = 100, bootstrap = True, random_state = 0) 
    
modelRF.fit(X_train, y_train)

scoreRF_ff = modelRF.score(X_validate,y_validate)

print('Random Forest classifier accuracy score with full dataset validation data and best parameters = ',scoreRF_ff)
```

```python
y_predict = modelRF.predict(X_validate)

cmRF_ff = confusion_matrix(y_validate, y_predict)

print('Random Forest classifier Confusion Matrix with full dataset validation data and best parameters = \n', cmRF_ff)
```

Support Vector Classifier (SVC)

```python
# Create an instance with SVC
from sklearn.svm import SVC
modelSVC = SVC(max_iter = 1000) 

modelSVC.fit(X_train, y_train)

modelSVC.score(X_test,y_test)
```

```python
# Optimize a SVC Model using GridSearch

param_grid = {'C': [0.1,1, 10, 100, 1000],
              'gamma': [1,0.1,0.01,0.001,0.0001],
              'kernel': ['linear','rbf']} 
```

```python
# import GridSearchCV from model_selection
from sklearn.model_selection import GridSearchCV

grid = GridSearchCV(SVC(), param_grid, verbose=2)
grid.fit(X_train,y_train)
```

```python
print('The best parameters are %s with a score of %0.2f' 
      % (grid.best_params_, grid.best_score_))
```

```python
# Validate the model with best parameters and validation data. Then calculate confusion matrix
modelSVC = SVC(C = 10, gamma = 0.01, kernel = 'rbf', max_iter = 1000) 

modelSVC.fit(X_train, y_train)

scoreSVC_ff = modelRF.score(X_validate,y_validate)

print('SVC accuracy score with full dataset validation data and best parameters = ',scoreSVC_ff)
```

```python
y_predict = modelSVC.predict(X_validate)

cmSVC_ff = confusion_matrix(y_validate, y_predict)

print('SVC Confusion Matrix with full dataset validation data and best parameters = \n', cmSVC_ff)
```
