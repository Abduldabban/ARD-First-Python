```python
from sklearn.preprocessing import LabelEncoder
df['diagnosis'] = LabelEncoder().fit_transform(df['diagnosis'])
df
```

```python
plt.title('Diagnosis Histogram')
plt.xlabel('diagnosis')
plt.ylabel('Number of Readings')
plt.hist(df['diagnosis'], bins =2)
```

```python
# assign attributues to X and y variables
y = df.diagnosis.values

features = len(selected_features_df.columns)
X = selected_features_df.iloc[:, 0:features].values
X
```

```python
# Sacling the features array becasue we have big variability between features values

from sklearn.preprocessing import StandardScaler

X_scaled = StandardScaler().fit_transform(X)
```

```python
df_data_refined = pd.DataFrame(X_scaled)

df_data_refined.columns = ['perimeter_mean', 'compactness_mean', 'concavity_mean', 'perimeter_se']



df_data_refined['diagnosis'] = df['diagnosis']

df_data_refined
```

```python
df_data_refined.to_csv('data_refined.csv')
```
